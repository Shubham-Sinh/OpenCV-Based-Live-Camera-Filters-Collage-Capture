```python
import cv2 as cv
import numpy as np
import os
```

```python


folder = "selfie"
os.makedirs(folder, exist_ok=True)

cam = cv.VideoCapture(0)
count = 1

# Colors (BGR)
blue = [255, 0, 0]
green = [0, 255, 0]
red = [0,0,240]

while True:

    ret, frame = cam.read()
    if not ret:
        break

    frame = cv.flip(frame, 1)

    # Resize for collage consistency
    frame = cv.resize(frame, (320, 240))

    # 🎨 Filters
    bg1 = np.full(frame.shape, blue, dtype=np.uint8)
    bg2 = np.full(frame.shape, green, dtype=np.uint8)
    bg3 = np.full(frame.shape, red, dtype=np.uint8)
    

    blue_filter = cv.addWeighted(frame, 0.8, bg1, 0.2, 0)
    green_filter = cv.addWeighted(frame, 0.8, bg2, 0.2, 0)
    warm_filter = cv.addWeighted(frame,0.85, bg3, 0.22, 0)

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    gray = cv.cvtColor(gray, cv.COLOR_GRAY2BGR)

    blur = cv.GaussianBlur(frame, (15, 15), 0)

    edges = cv.Canny(frame, 100, 200)
    edges = cv.cvtColor(edges, cv.COLOR_GRAY2BGR)

    # 🧩 Collage (2 rows × 3 columns)
    top_row = np.hstack((frame, blue_filter, green_filter))
    bottom_row = np.hstack((gray, blur, edges))
    collage = np.vstack((top_row, bottom_row))

    cv.imshow("natural", frame)

    cv.imshow("Blur", blur)
    cv.imshow("Cool_tone",blue_filter)
    cv.imshow("green_filt",green_filter)
    cv.imshow("warm_tone",warm_filter)
    cv.imshow("Highlight", edges)
    cv.imshow("old-school",gray)

    # 🖥️ Show collage
    cv.imshow("Multi Filter Collage", collage)

    key = cv.waitKey(1) & 0xFF
    if key == 13:
        subfolder = f"{folder}/set_{count}"
        
        os.makedirs(subfolder, exist_ok=True)
        cv.imwrite(f"{subfolder}/natural.png", frame)
        cv.imwrite(f"{subfolder}/blue.png", blue_filter)
        cv.imwrite(f"{subfolder}/green.png", green_filter)
        cv.imwrite(f"{subfolder}/warm.png", warm_filter)
        cv.imwrite(f"{subfolder}/gray.png", gray)
        cv.imwrite(f"{subfolder}/blur.png", blur)
        cv.imwrite(f"{subfolder}/edges.png", edges)
        cv.imwrite(f"{subfolder}/collage.png", collage)
        print(f"Saved set in {subfolder}")
        count += 1

    
cam.release()
cv.destroyAllWindows()
```
